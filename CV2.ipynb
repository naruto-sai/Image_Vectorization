```python
# Captruing the live feed and converting it into individual frames and using 'a' to stop feeding into the current folder 
# and move on to the next folders until all the folders are done
vi=cv2.VideoCapture(0)
count=0
for i in dirs:
    while True:
        count+=1
        su,img=vi.read()    
        cv2.imshow("video",img)
        cv2.imwrite(r"Python Class\Open_CV\{}\{}.jpg".format(i,count),img)
        if cv2.waitKey(1)==ord("a"):
            break
    cv2.destroyAllWindows()
```
